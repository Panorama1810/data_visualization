```python
# 提取部分2019~2020年数据
avg_expense2020 = avg_expense[['地区','2020年']]
avg_income2020 = avg_income[['地区','2020年']]
edum2019 = edum[['地区','2019年']]
eduf2019 = eduf[['地区','2019年']]
unmarriedm2019 = unmarriedm[['地区','2019年']]
unmarriedf2019 = unmarriedf[['地区','2019年']]
gdp2020 = gdp[['地区','2020年']]
```

```python
# 结婚离婚全国数据统计
marry_sum = df(marry.drop('地区', axis=1))
divorce_sum = df(divorce.drop('地区', axis=1))
marry_sum.loc['sum'] = marry_sum.apply(lambda x: x.sum())
divorce_sum.loc['sum'] = divorce_sum.apply(lambda x: x.sum())
marry_sum = np.array(round(marry_sum.loc['sum'],2))[::-1]
divorce_sum = np.array(round(divorce_sum.loc['sum'],2))[::-1]
```

```python
# 数据转换与数据配准
# 计算离结比并与粗离婚率拼接
m = marry.iloc[:,1:12]; d = divorce.iloc[:,1:12]
d_m = d/m
provinces = marry.iloc[:,0]
d_m = pd.concat([provinces,d_m],axis=1)
d_m2020 = d_m[['地区','2020年']]
d_m2020 = d_m2020.sort_values(by='2020年')
raw_divorce2020 = raw_divorce[['地区','2020年']]
data2020 = pd.merge(d_m2020, raw_divorce2020, left_on='地区', right_on='地区')
data2020.columns = ['地区','离结比','粗离婚率']
data2020['离结比'] = [x*100 for x in data2020['离结比'].values]
data2020['粗离婚率'] = [x*0.1 for x in data2020['粗离婚率'].values]
```

```python
# 离结比与人均可消费支出、人均可支配收入、GDP拼接
d_m_compare = pd.merge(d_m2020, avg_expense2020, left_on='地区', right_on='地区')
d_m_compare.columns = ['地区','离结比','人均可消费支出']
d_m_compare = pd.merge(d_m_compare, avg_income2020, left_on='地区', right_on='地区')
d_m_compare.columns = ['地区','离结比','人均可消费支出','人均可支配收入']
d_m_compare = pd.merge(d_m_compare, gdp2020, left_on='地区', right_on='地区')
d_m_compare.columns = ['地区','离结比','人均可消费支出','人均可支配收入','GDP']
d_m_compare['离结比'] = [x*100 for x in d_m_compare['离结比'].values]
```

```python
# 绘制地图与时间轮播图
def map_time() -> Timeline:

    t = Timeline(init_opts=opts.InitOpts(width="1200px", height="700px"))  # 定制主题
    for i in range(2011,2021):
        year = marry[['地区',str(i)+'年']]
        min_data=min(marry[str(i)+'年'])
        max_data=max(marry[str(i)+'年'])
        map_chart = (
            Map(init_opts=opts.InitOpts(width='300px', height='300px'))
            .add(
                series_name="",
                data_pair=[list(z) for z in zip(list(year['地区']),list(year[str(i)+'年']))],
                zoom=0.9,
                center=[119.5, 34.5],
                is_map_symbol_show=False,
                itemstyle_opts={
                      'normal': {
                          'shadowColor': 'rgba(0, 0, 0, .5)',  # 阴影颜色
                          'shadowBlur': 5,  # 阴影大小
                          'shadowOffsetY': 1,  # Y轴方向阴影偏移
                          'shadowOffsetX': 1,  # x轴方向阴影偏移
                          'borderColor': '#fff'
                      }
                }
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(
                    title="2011-2020年全国各省结婚登记数（单位：万对）",
                    subtitle="制作人：19123172 何育辰       数据来源：国家统计局",
                    pos_left="center",
                    pos_top="top",
                    title_textstyle_opts=opts.TextStyleOpts(
                        font_size=25, color='Black'
                    ),
                ),
                tooltip_opts=opts.TooltipOpts(
                    is_show=False,
                ),
                visualmap_opts=opts.VisualMapOpts(
                    is_calculable=True,
                    dimension=0,
                    pos_left="2%",
                    pos_bottom="20%",
                    textstyle_opts=opts.TextStyleOpts(color="black",font_size=15),
                    min_=min_data,
                    max_=max_data,
                ),
            )
        )
        
        grid_chart = (
            Grid()
            .add(map_chart, grid_opts=opts.GridOpts())
        )

        t.add(grid_chart,str(i))


    t.add_schema(
            orient="vertical",
            is_auto_play=True,
            is_inverse=True,
            play_interval=2000,
            pos_left="null",
            pos_right="0",
            pos_top="10",
            pos_bottom="10",
            width="60",
            label_opts=opts.LabelOpts(is_show=True, interval=0,color="Black",font_size = 15,font_weight="bold"),
    )
    
    rangecolor = ['DeepSkyBlue','Orange','Red']

    titlecolor='Black'
    
    return t
```
